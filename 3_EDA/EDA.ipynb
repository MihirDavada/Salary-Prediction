**Check The Data Is Normally Distributed Or Not....**

```python
numeric = ['Rating', 'average_salary', 'age_of_company', 'desc_len']
for i, col in enumerate(numeric):
    plt.figure(i)
    plt.hist(df[col])
    plt.title(f'Histogram for {col}')
plt.show()
```

**Rating And Average_salary is close to normally distributed but age_of_company has skewed distribution(exponential) if we want to perform regression, We Have To Convert It Into Normal Distribution**

```python
cat = ['Rating', 'Type of ownership', 'Sector', 'Revenue', 'job_state','job_simp', 'seniority', 'num_competitors']

for i , col in enumerate(cat):
    cat_num = df[col].value_counts()
    print(f" Graph For {col} -  total : {len(cat_num)} Categories")
    chart = sns.barplot(x = cat_num.index , y = cat_num )
    chart.set_xticklabels(chart.get_xticklabels(), rotation = 90)
    plt.show()
```

- **Insights**

    - Most Of The Companies's Rating Is Between  3.5 TO 4
    
    - Most Of The Companies Are Private.
    
    - Highest Job Openings For IT Sector
    
    - Most Of The Companies Revenue Is Greater Then 100 to 500 million(USD).
    
    - Most Of The Companies Are Located In California , New York And Massachusetts 
    
    - Highest Job Openings Is For Data Science Role.
    
    - Most Of The Companies Do Not Have Any competitors Or Some Compinies Have 3. 

```python
pd.pivot_table(df , index = 'job_simp' , values = 'average_salary')
```

```python
pd.pivot_table(df , index = ['job_simp', 'seniority'] , values = 'average_salary')
```

**pivot_table() creates group of job_state( CA , IL , DC ) and then average is calculated for each group**

```python
pd.pivot_table(df , index = ['job_state'] , values = 'average_salary').sort_values('average_salary', ascending = False )
```

```python
pd.pivot_table(df , index = ['job_state', 'job_simp'] , values = 'average_salary')
```

**pivot_table() Is As Same As groupby**

```python
df.groupby(['job_state','job_simp']).mean()['average_salary']
```

```python
pd.pivot_table(df , index = ['job_state', 'job_simp'] , values = 'average_salary', aggfunc = "count").sort_values('job_state', ascending = False )
```

**We Are Creating pivot_table For Each Column VS average_salary**

```python
pivot_li = ['Rating', 'job_state', 'Size','Type of ownership', 'Sector', 'Revenue', 'age_of_company', 'python', 'r_studio', 'spark', 'aws', 'excel', 
            'seniority', 'num_competitors'] 

for i , col in enumerate(pivot_li):
    print()
    print(f" Pivot Table For {i} --> {col} ")
    print()
    print(pd.pivot_table(df , index = col , values = 'average_salary').sort_values('average_salary', ascending = False ))
    
    
    
```

**Salary For Python Jobs**

```python
pivot_li = ['Rating', 'job_state', 'Size','Type of ownership', 'Sector', 'Revenue', 'age_of_company','seniority', 'num_competitors'] 

for i , col in enumerate(pivot_li):
    print()
    print(f" Pivot Table For {i} --> {col} ")
    print()
    print(pd.pivot_table(df , index = col , columns = 'python' , values='average_salary' , aggfunc = 'count'))
    
    
    
```
